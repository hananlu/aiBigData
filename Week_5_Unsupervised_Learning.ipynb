```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Membuat dataset
X, y = make_blobs(
    n_samples=200,
    centers=3,
    random_state=42
)
```

```python
# Membuat model
kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

# Training
kmeans.fit(X)
```

```python
from sklearn.metrics import silhouette_score

# Calculate Silhouette Score for n_clusters=3
silhouette_avg = silhouette_score(X, kmeans.labels_)

print(f"Silhouette Score for K-Means with 3 clusters: {silhouette_avg:.4f}")
```

```python
# Hasil cluster
labels = kmeans.labels_

# Visualisasi
plt.scatter(
    X[:,0],
    X[:,1],
    c=labels
)

plt.title("K-Means Clustering")

plt.show()
```

```python
inertia = []

K_range = range(1,10)

for k in K_range:

    kmeans = KMeans(n_clusters=k)

    kmeans.fit(X)

    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia)

plt.xlabel("Jumlah Cluster")

plt.ylabel("Inertia")

plt.title("Elbow Method")

plt.show()
```
