### Feature Engineering

```python
data = {
    "tanggal": ["2024-01-01", "2024-02-05", "2024-03-10"]
}

df = pd.DataFrame(data)

df["tanggal"] = pd.to_datetime(df["tanggal"])

df["bulan"] = df["tanggal"].dt.month

df["hari"] = df["tanggal"].dt.day
```
