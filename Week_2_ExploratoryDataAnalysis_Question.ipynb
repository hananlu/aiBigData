### Load Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {"Nama": ["BoB", "Alice", "Keith"],
        "Jenis Kelamin": ["Laki-Laki", "Perempuan", "Laki-Laki"],
        "Nilai":[70, 80, 90]}
```

```python
data = {
    "Name": [
        "Andi", "Budi", "Siti", "Rina", "Doni",
        "Lina", "Agus", "Maya", "Rudi", "Tono"
    ],
    "Age": [
        21, 22, 20, 23, 21,
        24, 22, 23, 25, 21
    ],
    "City": [
        "Jakarta", "Bandung", "Surabaya", "Jakarta", "Bandung",
        "Surabaya", "Jakarta", "Bandung", "Surabaya", "Jakarta"
    ],
    "Salary": [
        5000, 6000, 5500, 7000, 6500,
        7200, 5800, 6100, 6800, 5900
    ]
}


df = pd.DataFrame(data)
```

### View Distribution Values

```python
sns.displot(df["Salary"], kde=True)

plt.title("Distribusi Pendapatan")

plt.show()
```

### Scatter Plot (Age vs Salary)

```python
sns.scatterplot(x="Age", y="Salary", data=df)

plt.title("Perbandingan Umur dengan Pendapatan")

plt.show()
```

```python
df["Age"].corr(df["Salary"])
```

### Pie Plot City Distribution

```python
city_counts = df["City"].value_counts()

plt.pie(
    city_counts,
    labels=city_counts.index,
    autopct="%1.1f%%"
)

plt.title("City Distribution")

plt.show()
```

### Bar plot

```python
totalSalary = df.groupby("City")["Salary"].sum().reset_index()
totalSalary
```

```python
sns.barplot(x='City', y="Salary", data=totalSalary)

plt.title("Total Pendapatan untuk Setiap Kota")

plt.show()
```
